```python
import cv2, os
import matplotlib.pyplot as plt
from tqdm import tqdm

base_path = "/home/ssd_4tb/minjae/KITTI/dataset/sequences/07/image_0"

# ORB 객체 생성
orb = cv2.ORB_create()

# 이미지 로드
for img_path in tqdm(os.listdir(base_path)):
    img_abs_path = os.path.join(base_path, img_path)
    img = cv2.imread(img_abs_path, cv2.IMREAD_GRAYSCALE)

    # 키포인트 및 디스크립터 추출
    keypoints, descriptors = orb.detectAndCompute(img, None)
    print(descriptors.shape, keypoints)
    # 키포인트 시각화
    img_keypoints = cv2.drawKeypoints(img, keypoints, None, color=(0,255,0))

    cv2.imshow('Image', img_keypoints)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
